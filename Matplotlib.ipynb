```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
plt.plot([4,5,2],[3,5,6],'go-.',linewidth=2,markersize=5)
```

```python
x = np.linspace(-np.pi,np.pi,50); # np.linespace() içerisine ilk olarak başlangıç değeri, ikinci olarak bitiş değerini alır.
# En sondaki 50 ise bu başlangıç ve bitiş değerini 50 eşit parçaya böler ve bize bunu array olarak döndürür.
x
```

```python
plt.plot(x,np.sin(x),'go--',markersize=0,ls=':',linewidth=3,label='sinüs',alpha=0.3) # np.sin(x)=y değerleri olacak
plt.plot(x,np.cos(x),'ro--',markersize=0,ls=':',linewidth=3,label='cosinüs',alpha=1) # np.cos(x)=y değerleri olacak
plt.legend() # yukarıda label'ın içerisine yazılanları göstermek için kullanılır.
```

# Stiller ve Grafik Etiketlendirme

```python
plt.style.use('ggplot')
plt.plot(x,np.sin(x),'go--',markersize=0,ls=':',linewidth=3,label='sinüs',alpha=1) 
plt.plot(x,np.cos(x),'ro--',markersize=0,ls=':',linewidth=3,label='cosinüs',alpha=1) 
plt.legend()
# Grafiği kırpmak için aşağıdaki fonksiyonlar kullanılabilir.
#plt.xlim(-2,2) 
#plt.ylim(-1,1)
plt.axis("tight")
# Eksenleri isimlendirme
plt.title("Cosinüs ve Sinüs grafikleri")
plt.xlabel("x-ekseni")
plt.ylabel("y-ekseni")
```

# Bar grafiği
Kategorik verilerde tercih edilebilir.

```python
x1=np.random.randint(10,size=5)
y1=np.random.randint(10,size=5)
x2=np.random.randint(10,size=5)
y2=np.random.randint(10,size=5)
```

```python
x1,y1
```

```python
x2,y2
```

```python
plt.bar(x1,y1,label="x1,y1")
plt.bar(x2,y2,label="x2,y2")
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
```

```python
# Kategorik verilerde bar grafiği
x=np.arange(5) # 0-3 arası sıralı şekilde 3 dahil edilmeden sayı üretir.
plt.bar(x,height=[1,2,3,4,5],edgecolor='white',color='green') # y sınırını height ile biz belirledik.
plt.xticks(x,['Başarısız','Geçer','Orta','İyi','Pek iyi'])
# xticks() ile x eksenindekileri yeniden isimlendir diyip.Yeni isimleri giriyoruz.

plt.title("Karne rakam anlamları")
plt.xlabel("Anlamları")
plt.ylabel("Notlar")
```

# Histogram
Yoğunluk tabanlı bir grafiktir

```python
np.random.seed(12345) # Üretilen random sayı değerleri her run edildiğinde değerler değişmesin diye seedi böyle sabitliyoruz.
kadin_yaslari=np.random.randint(low=1,high=75,size=100)
erkek_yaslari=np.random.randint(low=1,high=75,size=100)
```

```python
plt.hist(kadin_yaslari,label="Kadın Yaşları")
plt.hist(erkek_yaslari,label="Erkek Yaşları")
plt.legend()
```

```python
np.random.seed(12345)
x=np.random.normal(loc=0,scale=1,size=100000)
plt.hist(x,density=True,bins=30); # density=True yaparsak verileri olasıksal sayılara (normalize değerlere) dönüştürür.
# bins ile bar çerçevelerini küçülterek daha görüntüyü daha toplu hale getirdik.
# ; kullanarak grafiğin üstünde yer alan arrayların gösterimini engelledik.
```

```python
ax.set_title("Başlık")
```

# Etiket ve legend ayarları

```python
fig=plt.figure()
ax=fig.add_subplot(1,1,1)
data1=np.random.randn(1000).cumsum()
data2=np.random.randn(1000).cumsum()
data3=np.random.randn(1000).cumsum()
ax.plot(data1,"k:",label="bir")
ax.plot(data2,"r--",label="iki")
ax.plot(data3,"g-.",label="üç")
ax.legend(loc="best") # grafikteki en iyi alana tabloyu koyar
```
