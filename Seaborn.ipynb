<b> 13, 18, 13, 14, 13, 16, 14, 21, 13 </b> sayılarımız olsun. <br>

# Matrix Plots

## Heatmaps-Isı Haritası
Verileri satır ve sütun olarak matrisler oluşturup birbirleri arasındaki ilişkiyi ısı haritaları ve korelasyon sayıları ile birlikte
oluşturulabiliriz. Ancak bazı veri setlerini satır sütun olarak bölsek bile korelasyon sayılarını kullanmamız çok anlam ifade etmez(Örneğin kategorik veriler ve korelasyon ilişkisi bize çok bir anlam ifade etmeyecektir).
Bu yüzden bu tür veri setleri için Pivot tablolar yardımı ile ısı haritaları oluşturabiliriz.

```python
plt.figure(figsize=(8,6))
sns.set_context('paper',font_scale=1.4)
# 1.YOL
crash_mx=crash_df.corr() # değikenlerin korelasyon sayılarını aldık. Isı haritasını korelasyon sayıları ile oluşturmak gerekir.
sns.heatmap(crash_mx,annot=True,cmap='Blues') # annot=True ile korelasyon değerlerini, cmap ile renk paletini ayarlardık.
```

```python
# 2.YOL - Pivot Tablolar
flights=sns.load_dataset('flights')
flights=flights.pivot_table(index='month',columns='year',values='passengers')
# Pivot table ile "month" değerlerini satırlara index olarak yerleştirdik, "year" değerlerini sütun olarak tanımladık.
# Ve değerleri "passengers" ise satır ve sütun arasında matris oluşturacak şekilde yerleştirdik.
```

```python
sns.heatmap(flights,cmap='Blues',linecolor='white',linewidths=1) # verileri line ile ayırıp renk ve ölçüsünü belirledik.
# Görüldüğü üzere uçuşların çoğu temmuz ve ağustos aylarında gerçekleşmekte ve insanlar 1949 a kıyasla 1960 da daha fazla uçmuş.
```

## Cluster Map-Küme Haritası

```python
iris=sns.load_dataset('iris')
species=iris.pop('species')
# Sadece tür kısmındaki değerleri alıp değişkene atadık. Çünkü string ifadeleri ısı haritasında kullanmak istemiyoruz.
sns.clustermap(iris,cmap='Blues',standard_scale=1)
```

## Heatmap ve Cluster Map arasındaki farklar
- Cluster Map, varsayılan olarak uzaklık matrisi elde etmek için öklid ölçüsü ve kümeleme için tam aglomerasyon yöntemini kullanırken , Heatmap sırasıyla korelasyon ve ortalama aglomerasyon yöntemini kullanır.

- Cluster Map, mesafe matrisini hesaplar ve ölçeklendirmeden önce kümeleme algoritmasını çalıştırır, oysa Heatmap önceden ölçeklenmiş verileri kümeler.

- Cluster Map, burada açıklandığı gibi satır ve sütun ortalama değerlerine göre dendrogramı yeniden sıralar.

# Pair Grid

```python
## .map() fonksiyonunun işlevi
- <b>map()</b> fonksiyonu içerisine aldığı parametreyi bağlandığı (* değişken_adı.map(x) veya ** değiken_adı.map(x,y)) yapıya ya da içerisinde aldığı fonksiyon-parametrelere göre işlemler yapar.
- <b>sonuc = map(karesi,liste)</b> Burada <b>karesi</b> adındaki fonksiyona gider bu fonskiyonun içerisine değişken adı <b> liste </b> olan değişkenin içerisindeki elemanları <b>karesi</b> fonskiyonunda işleme sokar.
- <b>map(lambda x:x ** 2,liste)</b> Burada ise yukarıdaki <b>karesi</b> fonksiyonun yaptığı işi map() fonskiyonunun kendisi içerisinde bulunan <b>lambda fonksiyonu</b> ile yapıyoruz.
```

```python
sns.set_style('darkgrid')
iris2=sns.load_dataset('iris')
iris_g=sns.PairGrid(iris2,hue='species',hue_kws={"marker": ["D", "s", "o"]},palette='Set2')
# Scatter plot için markerlerı belirledik.
#iris_g.map(sns.scatterplot)
iris_g.map_diag(plt.hist) # Yalnızca köşegene histogram grafiği çizdirir.
#iris_g.map_offdiag(sns.scatterplot) # Köşegen dışındaki alanlara scatter grafiği çizdirir.
iris_g.map_upper(sns.scatterplot,edgecolor="k", s=20) # Köşegenin üstündeki alana scatter plot çizdirir.
iris_g.map_lower(sns.kdeplot) # Köşegenin altındaki alana scatter plot çizdirir.
iris_g.add_legend()
```

```python
iris_g=sns.PairGrid(iris2,hue='species',hue_kws={"marker": ["D", "s", "o"]},palette='Accent',
                    x_vars=["sepal_length", "sepal_width"],y_vars=["petal_length", "petal_width"])
iris_g.map(sns.scatterplot) 
iris_g.add_legend()
```
